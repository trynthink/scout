### FIND DIFFERENCE BETWEEN 2024 and 2023 FACTORS

```python
## Absolute difference
import pandas as pd

filedir = "./output/"
# filename = "Com_Cdiv_EMM_amy2018.csv"


filenames = ["Com_Cdiv_State_amy2018.csv", "Com_Cdiv_EMM_amy2018.csv",
             "Res_Cdiv_EMM_amy2018.csv", "Res_Cdiv_State_amy2018.csv"]
             
for filename in filenames:
    file1_path = f"{filedir}2023/{filename}"
    file2_path = f"{filedir}2024_end_use/{filename}"
    df1 = pd.read_csv(file1_path)
    df2 = pd.read_csv(file2_path)

    sorted_df1 = df1.sort_values(by=["End use", "CDIV"]).reset_index(drop=True)
    sorted_df2 = df2.sort_values(by=["End use", "CDIV"]).reset_index(drop=True)
    # if filename =="Com_Cdiv_State_amy2018.csv":
    #     print(filename)
    #     print(sorted_df1.iloc[0:9])
    #     print(sorted_df2.iloc[0:9])
    
    common_columns = sorted_df1.columns.intersection(sorted_df2.columns)
    df1_common = sorted_df1[common_columns]
    df2_common = sorted_df2[common_columns]
    df1_numeric = df1_common.select_dtypes(include='number')
    df2_numeric = df2_common.select_dtypes(include='number')
    absolute_difference_df = df2_numeric.subtract(df1_numeric, fill_value=0)
    absolute_difference_df_from_third_column = absolute_difference_df.iloc[:, 1:]
    # percentage_difference_df = df1_numeric.subtract(df2_numeric, fill_value=0).abs().divide(df1_numeric, fill_value=1)
    # percentage_difference_df_from_third_column = percentage_difference_df.iloc[:, 1:]

    shifted_absolute_difference_df = pd.concat([pd.DataFrame(index=df1_common.index, columns=["placeholder"]), absolute_difference_df_from_third_column], axis=1)
    df_with_inserted_columns = pd.concat([df1_common.iloc[:, :2], shifted_absolute_difference_df], axis=1)

    df_with_first_two_columns = pd.concat([df1_common.iloc[:, :2], absolute_difference_df_from_third_column], axis=1)
    # df_with_first_two_columns = pd.concat([df1_common.iloc[:, :2], percentage_difference_df_from_third_column], axis=1)
    
    df_with_first_two_columns = df_with_first_two_columns.fillna(0)
    output_path = f"{filedir}pctdiff/absdiff_{filename}"
    df_with_first_two_columns.to_csv(output_path, index=False)
    print("COMPLETE")

# tools.display_dataframe_to_user(name="Absolute Differences (Starting from Third Column)", dataframe=absolute_difference_df_from_third_column)
# # Output path for reference
# output_path



#### ResStock, number of homes; ComStock, square footage
```
